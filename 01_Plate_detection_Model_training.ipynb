### Downloading Pretrained Yolov8 Model

# Saving the model

```python
from ultralytics import YOLO

# Path to your trained weights
weights_path = "/content/runs/detect/train/weights/best.pt"

# Load the model with the trained weights
model = YOLO(weights_path)

# Now you can use this model for inference
# For example, to run inference on an image:
results = model("/content/drive/MyDrive/soulpage/images/test/1000.jpg")

# To save the model to your Google Drive
import os
from google.colab import drive

# Mount Google Drive if not already mounted
drive.mount('/content/drive')

# Define the save path in your Google Drive
save_path = "/content/drive/MyDrive/your_model_folder/best_model.pt"

# Ensure the directory exists
os.makedirs(os.path.dirname(save_path), exist_ok=True)

# Save the model
model.save(save_path)

print(f"Model saved to {save_path}")
```

### Detacting the License plate using pretrained model

```python
from PIL import Image
import matplotlib.pyplot as plt

# Assuming you've run inference on an image
results = model("/content/drive/MyDrive/soulpage/images/test/1000.jpg")

# Plot the results
plt.figure(figsize=(10, 10))
plt.imshow(results[0].plot())
plt.axis('off')
plt.show()
```

```python
# Print the results
for r in results:
    print(r.boxes)  # print detection boxes
    print(r.boxes.xyxy)  # print box coordinates
    print(r.boxes.conf)  # print confidence scores
    print(r.boxes.cls)   # print class labels
```
